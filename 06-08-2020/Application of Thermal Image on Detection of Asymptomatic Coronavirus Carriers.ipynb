## 背景

根据[Wang et al.](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC5780734/pdf/bmjopen-2017-017964.pdf)，肺部红外图像（Chest Thermal Image）在诊断肺炎时具有和肺部X光片（Chest X-Ray）类似的效果。文章指出，在肺部发炎区域会有相较于其他肺部区域异常的温度，由此可以大致确定病灶位置并作出是否患有肺炎的判断。

这一结论既为我们的工作提供了可靠的思路和方向，同时也提出了新的挑战。

### 医学证据支持
首先，这项研究的focal pneumonia是一个会在一个肺部区域里集中产生炎症的肺炎。相对应的，viral pneumonias就是一种会扩散而非集中于一点的肺炎。因此就不会像focal pneumonia一样在红外图像中显示出局部的异常温度。所以，我们在选择这篇文章作为出发点时，需要获得更多的证据显示新冠肺炎在早期在肺部产生病灶有集中的炎症（inflammation）表现。

### 研究方法
其次，在获得了可靠的肺部区域之后，有两种潜在的方式可以对是否有发热病灶进行判断。

**传统**的做法是将选中的肺部区域每一个像素的RGB值做成一个分布图（distribution），依据之前提到的异常温度，相对应的红色（red）通道就会有异常的分布。接下来计划在这一方面进行一个初步的研究，探索正常分布（无症状普通人）和异常分布（无症状感染者）之间的是否有显著差异。同时观察显著差异是否具有普遍性。

**前沿**的做法是使用深度学习模型，在学习大量无症状普通人和无症状感染者肺部红外图像之后，使用分类器来对是否是潜在的无症状感染者做出判断。由于是在移动端部署，需要使用轻量级网络模型，所以还需要更多的研究和尝试才能得知深度学习模型在达到实时视频处理的同时，是否能够达到令人满意的准确率。

### 参考
Wang LT, Cleveland RH, Binder W, et al. Similarity of chest X-ray and thermal imaging of focal pneumonia: a randomised proof of concept study at a large urban teaching hospital. BMJ Open. 2018; 8(1): e017964. doi:10.1136/bmjopen-2017-017964

### RGB直方图

#### 范例1
在以下代码中，首先从肺部分割深度学习模型的结果中获取肺部的坐标。其次使用所得的肺部坐标制作一个覆盖层。然后使用OpenCV中的calcHist方程，利用之前制作的覆盖层，分析得出目标区域（肺部）的RGB直方图，并使用matplotlib展示。

参考: https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_imgproc/py_histograms/py_histogram_begins/py_histogram_begins.html#

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(20,10))

img = cv2.imread("1247.jpg")
test_img = cv2.imread("test.jpg")

mask = np.zeros(img.shape[:2], np.uint8)
mask[106:156, 296:342] = 255 #从肺部分割深度学习模型的结果获取的肺部坐标
masked_img = cv2.bitwise_and(img, img, mask=mask)

plt.subplot(221).set_title("Original Image", size=20), plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(222).set_title("Image w. Lung Selection", size=20), plt.imshow(cv2.cvtColor(test_img, cv2.COLOR_BGR2RGB))
plt.subplot(223).set_title("Masked Image", size=20), plt.imshow(cv2.cvtColor(masked_img, cv2.COLOR_BGR2RGB))

plt.subplot(224).set_title("RGB Histogram of Masked Image (Lung Selection)", size=20)
color = ('b','g','r')
for i,col in enumerate(color):
    hist = cv2.calcHist([img], [i], mask, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.show()
```

#### 范例2
在以下代码中，展示了肺炎患者的肺部热成像图片的RGB直方图。原热成像图片来自背景中的论文。

从以下结果中很难得出任何有力结论，还需要将无肺炎人的肺部热成像图片的RGB直方图与上图进行对比。

```python
plt.figure(figsize=(20,10))

original = cv2.imread("original.jpg")
cropped = cv2.imread("thermal.jpg")
plt.subplot(221).set_title("Original Thermal Image", size=20), plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
plt.subplot(222).set_title("Cropped Thermal Image", size=20), plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))

plt.subplot(223).set_title("RGB Histogram of Thermal Image", size=20)
color = ('b','g','r')
for i,col in enumerate(color):
    hist = cv2.calcHist([cropped], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([0,256])

plt.subplot(224).set_title("RGB Histogram of Thermal Image (Zoomed)", size=20)
color = ('b','g','r')
for i,col in enumerate(color):
    hist = cv2.calcHist([cropped], [i], None, [256], [0, 256])
    plt.plot(hist, color=col)
    plt.xlim([200,256])

plt.show()
```

#### 范例3
在以下结果中可以观察到，通过使用白色rgb色域的mask，可以将白色以及近似白色的区域分割出来，方便进一步的研究。

```python
plt.figure(figsize=(20,10))
plt.subplot(121).set_title("Cropped Thermal Image", size=20), plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))

plt.subplot(122).set_title("Cropped Thermal Image with White Mask", size=20)
lower = np.uint8([180, 180, 180])
upper = np.uint8([255, 255, 255])
white_mask = cv2.inRange(cropped, lower, upper)
masked_cropped = cv2.bitwise_and(cropped, cropped, mask=white_mask)

plt.imshow(cv2.cvtColor(masked_cropped, cv2.COLOR_BGR2RGB))
plt.show()
```
